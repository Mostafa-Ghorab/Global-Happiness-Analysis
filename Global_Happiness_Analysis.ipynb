```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
region_mapping = {
    'Western Europe': ['Switzerland', 'Iceland', 'Denmark', 'Norway', 'Finland', 'Netherlands', 'Sweden'],
    'North America': ['United States', 'Canada'],
    'Australia and New Zealand': ['New Zealand', 'Australia'],
    'Middle East and Northern Africa': ['Israel', 'United Arab Emirates', 'Oman', 'Qatar', 'Saudi Arabia', 'Bahrain', 'Kuwait', 'Jordan', 'Turkey'],
    'Latin America and Caribbean': ['Costa Rica', 'Mexico', 'Brazil', 'Panama', 'Chile', 'Argentina', 'Uruguay', 'Colombia'],
    'Southeastern Asia': ['Thailand', 'Singapore', 'Malaysia', 'Vietnam', 'Indonesia', 'Philippines'],
    'Central and Eastern Europe': ['Czech Republic', 'Slovakia', 'Poland', 'Hungary'],
    'Eastern Asia': ['Japan', 'South Korea', 'Taiwan'],
    'Sub-Saharan Africa': ['Nigeria', 'Ethiopia', 'Kenya', 'Ghana', 'Zimbabwe', 'Sierra Leone', 'Mauritania'],
    'Southern Asia': ['Bhutan', 'Pakistan', 'India', 'Nepal', 'Sri Lanka'],
}

df_combined['Region'] = df_combined['Country'].map({country: region for region, countries in region_mapping.items() for country in countries})

average_happiness_rank_by_region = df_combined.groupby('Region')['Happiness Rank'].mean()
average_happiness_score_by_region = df_combined.groupby('Region')['Happiness Score'].mean()

comparison_df = pd.DataFrame({
    'Average Happiness Rank': average_happiness_rank_by_region,
    'Average Happiness Score': average_happiness_score_by_region
})

comparison_df = comparison_df.sort_values(by='Average Happiness Rank')

fig, ax1 = plt.subplots(figsize=(12, 8))

color = 'tab:blue'
ax1.set_xlabel('Region')
ax1.set_ylabel('Average Happiness Rank', color=color)
ax1.plot(comparison_df.index, comparison_df['Average Happiness Rank'], marker='o', linestyle='-', color=color, label='Average Happiness Rank')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticks(np.arange(len(comparison_df)))
ax1.set_xticklabels(comparison_df.index, rotation=45, ha='right')

ax2 = ax1.twinx()
color = 'tab:grey'
ax2.set_ylabel('Average Happiness Score', color=color)
ax2.plot(comparison_df.index, comparison_df['Average Happiness Score'], marker='o', linestyle='-', color=color, label='Average Happiness Score')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.title('Comparing Regional Happiness Ranks and Scores')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

mean_happiness_rank = df_combined.groupby('Country')['Happiness Rank'].mean().sort_values()

top_15_countries = mean_happiness_rank.head(15)

ranks = range(1, 16)

sizes = (top_15_countries.values.max() - top_15_countries.values) + top_15_countries.values.min()

colors = ['#084594', '#2171B5', '#4292C6', '#6BAED6', '#9ECAE1', '#FFFFFF']

num_colors = len(top_15_countries)
color_map = colors[:num_colors] if num_colors <= len(colors) else plt.cm.Blues(np.linspace(0.2, 0.8, num_colors))

plt.figure(figsize=(12, 8))

squarify.plot(sizes=sizes, label=[f'{label}\nRank: {rank}'
                                  for label, rank in zip(top_15_countries.index, ranks)], color=color_map, alpha=0.8)

plt.axis('off')

plt.title('Top 15 Happiest Countries by Average Happiness Rank (Treemap)', fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

mean_happiness_rank = df_combined.groupby('Country')['Happiness Rank'].mean().sort_values()

bottom_15_countries = mean_happiness_rank.tail(15)[::-1]

bottom_15_countries_ranked = bottom_15_countries.rank(method='first', ascending=False).astype(int)

sizes = (bottom_15_countries.values.max() - bottom_15_countries.values) + bottom_15_countries.values.min()

colors = ['#084594', '#2171B5', '#4292C6', '#6BAED6', '#9ECAE1', '#C6DBEF', '#FFFFFF']

num_colors = len(bottom_15_countries)
color_map = colors[:num_colors] if num_colors <= len(colors) else plt.cm.Blues(np.linspace(0.2, 0.8, num_colors))

plt.figure(figsize=(12, 8))

squarify.plot(sizes=sizes, label=[f'{label}\nRank: {bottom_15_countries_ranked[label]}'
                                  for label in bottom_15_countries.index], color=color_map, alpha=0.8)

plt.axis('off')

plt.title('Bottom 15 Least Happy Countries by Average Happiness Rank (Treemap)', fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

columns_of_interest = ['Happiness Rank', 'Economy (GDP per Capita)', 'Family',
                       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
                       'Generosity', 'Dystopia Residual']

available_columns = [col for col in columns_of_interest if col in df_combined.columns]

if len(available_columns) != len(columns_of_interest):
    missing_columns = set(columns_of_interest) - set(available_columns)
    print(f"Warning: The following columns are missing in df_combined: {missing_columns}")

df_subset = df_combined[available_columns]
correlation_matrix = df_subset.corr()
happiness_rank_correlations = correlation_matrix['Happiness Rank'].drop('Happiness Rank')
correlation_df = pd.DataFrame(happiness_rank_correlations)
correlation_df.columns = ['Correlation with Happiness Rank']
correlation_df = correlation_df.sort_values(by='Correlation with Happiness Rank')

print("Correlation coefficients between 'Happiness Rank' and other variables:")
print(correlation_df.round(2))

plt.figure(figsize=(10, 6))
correlation_df.plot(kind='bar', legend=None)
plt.title("Correlation with Happiness Rank")
plt.xlabel("Variables")
plt.ylabel("Correlation Coefficient")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
